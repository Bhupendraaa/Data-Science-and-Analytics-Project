```python
import numpy as np

def Si_ze(column):
  if isinstance(column, (int, float)):
    return column
  s_column = str(column)
  if s_column.endswith('M'):
    try:
      # Convert MB to KB
      return float(s_column.replace('M', '')) * 1024
    except ValueError:
      return np.nan
  elif s_column.endswith('k'):
    try:
      # Already in KB
      return float(s_column.replace('k', ''))
    except ValueError:
      return np.nan
  elif s_column == 'Varies with device':
    return np.nan # Represent unknown sizes as NaN
  else:
    try:
      return float(s_column)
    except ValueError:
      return np.nan

df['Size'] = df['Size'].apply(Si_ze)
```
